```python
#sum hourly data to get daily data
import pylab as plt
import datetime as datetime

a['ob_end_time'] = pd.to_datetime(a.ob_end_time, format='%Y %m %d %H:%M')
a['timestamp'] = a['ob_end_time'].dt.strftime('%m %d')
daily_suma = a.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

b['ob_end_time'] = pd.to_datetime(b.ob_end_time, format='%Y %m %d %H:%M')
b['timestamp'] = b['ob_end_time'].dt.strftime('%m %d')
daily_sumb = b.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

c['ob_end_time'] = pd.to_datetime(c.ob_end_time, format='%Y %m %d %H:%M')
c['timestamp'] = c['ob_end_time'].dt.strftime('%m %d')
daily_sumc = c.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

d['ob_end_time'] = pd.to_datetime(d.ob_end_time, format='%Y %m %d %H:%M')
d['timestamp'] = d['ob_end_time'].dt.strftime('%m %d')
daily_sumd = d.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

e['ob_end_time'] = pd.to_datetime(e.ob_end_time, format='%Y %m %d %H:%M')
e['timestamp'] = e['ob_end_time'].dt.strftime('%m %d')
daily_sume = e.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

f['ob_end_time'] = pd.to_datetime(f.ob_end_time, format='%Y %m %d %H:%M')
f['timestamp'] = f['ob_end_time'].dt.strftime('%m %d')
daily_sumf = f.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

g['ob_end_time'] = pd.to_datetime(g.ob_end_time, format='%Y %m %d %H:%M')
g['timestamp'] = g['ob_end_time'].dt.strftime('%m %d')
daily_sumg = g.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

h['ob_end_time'] = pd.to_datetime(h.ob_end_time, format='%Y %m %d %H:%M')
h['timestamp'] = h['ob_end_time'].dt.strftime('%m %d')
daily_sumh = h.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

i['ob_end_time'] = pd.to_datetime(i.ob_end_time, format='%Y %m %d %H:%M')
i['timestamp'] = i['ob_end_time'].dt.strftime('%m %d')
daily_sumi = i.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()

j['ob_end_time'] = pd.to_datetime(j.ob_end_time, format='%Y %m %d %H:%M')
j['timestamp'] = j['ob_end_time'].dt.strftime('%m %d')
daily_sumj = j.groupby('timestamp')['glbl_irad_amt'].sum().reset_index()



```
